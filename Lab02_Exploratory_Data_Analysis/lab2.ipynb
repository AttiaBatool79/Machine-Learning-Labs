 **BASIC PLOTTING...**

 ***1. BOX PLOT..
 Boxplot helps detect outliers (data points that are significantly different from others) 
and shows the distribution's spread, median, and quartiles.***

```python
sns.boxplot(x=d['Fare'])
plt.title("plot of fare")
plt.xlabel(("fare"))
plt.show()
```

***2. DIST PLOT...***
 Combines a histogram and a KDE curve to visualize the distribution of the numerical 
 data.
 • Identifies skewness, modality, and overall spread of the data

```python
sns.histplot(d['Age'], kde=True, bins=20, color='pink')
plt.title("Age Distribution with KDE")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

 ***3. HEATMAP PLOT..***
 shows pairwise correlations between numerical variables.
 Highlights strong positive or negative relationships for further analysis.

```python
sns.heatmap(d.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

 ***4. HISTOGRAM..***
 Visualizes the frequency distribution of Age.
 Useful for identifying common age ranges and trends in passenger demographics

```python
d['Age'].plot(kind='hist', bins=20, color='green', alpha=0.7)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

 ***5. PIECHART..***
 Highlights proportions or percentages of passengers who survived versus those who did not.

```python
survival_counts = d['Survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'skyblue'])
plt.title("Survival Proportion")
plt.show()
```

# Task 2: Numerical Data Analysis

2. **Question:** What are the key statistical properties and distributions of the numerical columns?  
   - Analyze columns such as `Age`, `Fare`, and `Parch` using summary statistics and visualizations.  
   - Plot histograms, boxplots, and KDE plots to understand distributions and detect outliers.


**1.Summary Statistics:**

```python
numerical_columns = ['Age', 'Fare', 'Parch']

# Summary statistics
print("Summary Statistics:")
print(df[numerical_columns].describe())

# Check for missing values in numerical columns
print("\nMissing Values:")
print(df[numerical_columns].isnull().sum())
```

**2.Visualizations:**

```python
# Visualize distributions
for col in numerical_columns:
    plt.figure(figsize=(14, 5))
    
    # Histogram with KDE plot
    plt.subplot(1, 3, 1)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Histogram and KDE Plot of {col}")
    
    # Boxplot
    plt.subplot(1, 3, 2)
    sns.boxplot(x=df[col], color='orange')
    plt.title(f"Boxplot of {col}")
    
    # KDE plot separately
    plt.subplot(1, 3, 3)
    sns.kdeplot(df[col], color='green')
    plt.title(f"KDE Plot of {col}")
    
    plt.tight_layout()
    plt.show()
```

# Task 3: Categorical Data Analysis

3. **Question:** What is the frequency distribution of categorical columns?  
   - Explore `Pclass`, `Sex`, `Embarked`, and `Survived` using `value_counts()`.  
   - Visualize the distributions using countplots and pie charts.


**1.Frequency Distribution:**

```python
# Categorical columns to analyze
categorical_columns = ['Pclass', 'Sex', 'Embarked', 'Survived']

# Frequency distribution using value_counts()
for col in categorical_columns:
    print(f"\nFrequency distribution of {col}:")
    print(df[col].value_counts())
```

**2.Visualizations:**

```python
# Visualizations
for col in categorical_columns:
    plt.figure(figsize=(14, 6))
    
    # Countplot
    plt.subplot(1, 2, 1)
    sns.countplot(data=df, x=col, palette='viridis')
    plt.title(f"Countplot of {col}")
    
    # Pie chart
    plt.subplot(1, 2, 2)
    df[col].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis'))
    plt.title(f"Pie Chart of {col}")
    plt.ylabel("")  # Removes y-axis label for pie chart
    
    plt.tight_layout()
    plt.show()
```

# Task 4: Relationship Between Variables (Bivariate Analysis)

4. **Question:** How are variables related to each other?  
   Analyze the relationship between:  
   - **Numerical-Numerical:** `Age` vs `Fare` (scatter plot and correlation).  
   - **Categorical-Categorical:** `Pclass` vs `Survived` (grouped bar chart).  
   - **Numerical-Categorical:** `Age` distribution across survival status (boxplot).


**1.Numerical-Numerical: Age vs Fare:**

```python
## Numerical-Numerical: Age vs Fare
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm', alpha=0.7)
plt.title('Scatter Plot of Age vs Fare (Colored by Survival Status)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

**2.Categorical-Categorical: Pclass vs Survived:**

```python
## Categorical-Categorical: Pclass vs Survived
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='viridis')
plt.title('Grouped Bar Chart of Pclass vs Survival Status')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['Did Not Survive', 'Survived'])
plt.show()
```

**3.Numerical-Categorical: Age distribution across survival status**

```python
## Numerical-Categorical: Age distribution across survival status
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Survived', y='Age', palette='Set2')
plt.title('Boxplot of Age Distribution by Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Age')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

# Task 5: Multivariate Analysis

5. **Question:** How do multiple variables interact with each other?  
   - Perform a pairwise analysis on selected columns such as `Age`, `Fare`, `Pclass`, and `Survived` using pairplots.  
   - Visualize overall correlations using a heatmap.


**1.Pairwise Analysis (Pairplot):**

```python
## Pairwise Analysis: Pairplot
selected_columns = ['Age', 'Fare', 'Pclass', 'Survived']
sns.pairplot(df[selected_columns], hue='Survived', palette='coolwarm', diag_kind='kde')
plt.suptitle('Pairplot of Age, Fare, Pclass, and Survived', y=1.02)
plt.show()
```

**2.Correlation Heatmap:**

```python
## Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df[selected_columns].corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Heatmap of Correlations Among Selected Variables')
plt.show()
```
